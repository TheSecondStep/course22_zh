## 保存猫狗分类模型

这是一个最简示例，展示如何在 Kaggle 上训练 fastai 模型，并保存以便在您的应用中使用。

```python
# Make sure we've got the latest version of fastai:
!pip install -Uqq fastai
```

首先，从 fastai 导入我们需要的所有内容：

```python
from fastai.vision.all import *
```

下载并解压我们的数据集，其中包含狗和猫的图片：

```python
path = untar_data(URLs.PETS)/'images'
```

我们需要一种方法来将图像标记为狗或猫。在这个数据集中，猫的图片被赋予了以大写字母开头的文件名：

```python
def is_cat(x): return x[0].isupper() 
```

现在我们可以创建我们的 `DataLoaders`：

```python
dls = ImageDataLoaders.from_name_func('.',
    get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat,
    item_tfms=Resize(192))
```

……并训练我们的模型，一个 resnet18（以保持其小巧和快速）：

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

现在我们可以导出训练好的 `Learner`。它包含运行模型所需的所有信息：

```python
learn.export('model.pkl')
```

最后，如果右侧的 Kaggle 侧边栏尚未打开，请将其打开，找到标记为"Output"的部分。打开 `/kaggle/working` 文件夹，你会看到 `model.pkl`。点击它，然后点击右侧出现的菜单，选择"Download"。几秒钟后，你的模型将被下载到你的电脑上，然后你就可以创建使用该模型的应用程序了。
